### Постройте круговую диаграмму, которая изображает состав отчисления студентов ИВТ-б-о-19-1
Воронкин Р.А. - 30 %

Сагдеев К.М. - 60 %

Гайчук Д.В. - 10 %

```python
import matplotlib.pyplot as plt
labels = ['Воронкин Р.А.','Сагдеев К.М.','Гайчук Д.В.']
values = [30,60,10]
colors = ['yellow','red','blue']
plt.pie(values,labels=labels,colors=colors)
plt.axis('equal')
plt.show()
```
