ЗАДАЧА:
С противоположных концов города длиной 240 км едут навстречу друг другу
Джамалудин и Виталя.Виталя начинает движение на 2 часа позже Джамалудина. Через сколько часов они встретятся, если скорость Джамалудина – 48 км/час, так как джамал нищий и едет на троллейбусе, а Виталя богатый и едет на такси, а скорость такси – 72 км/час? 

РЕШЕНИЕ:
Пусть х ч.– время движения второго мальчика, тогда время первого – (х+2) ч. Составим уравнение, учитывая, что сумма расстояний равна 240 км:72(х + 2) + 48х = 240 решив которое, получим, что х= 2.9ч. Тогда время, через которое они встретятся – 4.9 ч.

```python
import matplotlib.pyplot as plt
plt.plot([240, 200, 160, 120, 100, 80, 60, 40, 20, 10])
plt.plot([0, 5, 10, 15, 18, 20, 25])
plt.ylabel('x, км', fontsize=12, color='blue')
plt.xlabel('t, ч', fontsize=12, color='blue')
plt.text(4.8, 270, '')
plt.text(4.8, 35, '')
plt.grid()
plt.show()
```
